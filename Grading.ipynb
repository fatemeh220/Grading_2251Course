```python
import pandas as pd

Student_OWL = pd.read_csv("list_student_OWL.csv")
Student_OWL = Student_OWL.iloc[:,:3]
Student_OWL.head()
```

```python
Deadline = '2023-01-27'
Assignment_Repo_URL_End = '/blob/main/Lab2.py'
Student_Github=pd.read_csv("lab2-grades-1675208605.csv")
Assignment_Name =Student_Github['assignment_name'][0]
Student_Github.head()

Student_Github = Student_Github[['roster_identifier','submission_timestamp','points_awarded','student_repository_url']]
Student_Github.columns = ['Student ID', 'Time','Grade','Link']
Student_Github.head()

Student_OWL = Student_OWL.merge(Student_Github, how='left')
Student_OWL.head()
```

```python
for i in Student_OWL.index:
  if Student_OWL['Grade'][i]==0:
    print('The link is: '+ str(Student_OWL['Link'][i])+ Assignment_Repo_URL_End)
    Student_OWL['Grade'][i] = input('Please enter the updated grade for %s: ' %(Student_OWL['Student ID'][i]))
```

```python
from datetime import datetime
import numpy as np
pd.options.mode.chained_assignment = None  # default='warn'

Student_OWL['Delay'] = 0
Student_OWL['Grade_initial'] = Student_OWL['Grade']

for i in range(Student_OWL.shape[0]):
  # if(math.isnan(pd.to_numeric(Student_OWL['Time'][i]))):
  #   i=i+1
  try:
    res = datetime.strptime(Deadline, '%Y-%m-%d') - datetime.strptime(Student_OWL['Time'][i][:10], '%Y-%m-%d')
    number_of_days = res.days
    # res = datetime.strptime(datetime.strptime('2023-01-20', '%Y-%m-%d') - datetime.strptime(Student_OWL['Time'][i][:10], '%Y-%m-%d'))
    # number_of_days = res.days
    if(number_of_days<0):
      Student_OWL.Grade[i]=Student_OWL.Grade[i]*(1+(number_of_days)*0.1)
      Student_OWL['Delay'][i] = -number_of_days

  except:
  #   # i=i+1
    Student_OWL.Grade[i] = 0
 
Student_OWL.head()
```

```python
Student_OWL = Student_OWL[['Student ID', 'Name', 'Grade']]
Student_OWL.rename(columns = {'Grade':Assignment_Name+' [10]'}, inplace = True)
Student_OWL.head()
Student_OWL.to_csv(Assignment_Name+'.csv', index=False)
```
